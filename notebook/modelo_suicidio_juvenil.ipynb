## 🧠 Modelado Predictivo: Riesgo de Suicidio en Infancia y Adolescencia
En esta sección se construye un modelo de clasificación que intenta predecir si un caso puede ser considerado de "Riesgo Alto", según una definición basada en criterios psicosociales observados en el análisis exploratorio.


```python
# Asegurarte de trabajar sobre una copia, no una vista
df = df_infancia_adolescencia.copy()

# Ahora sí, crea la nueva columna sin advertencia
df.loc[:, "Riesgo Alto"] = 0
# Aplica las condiciones de riesgo
condicion_riesgo_alto = (
    (df["Ciclo Vital"] == '(12 a 17) Adolescencia') &
    (df["Razón del Suicidio"].isin(["Desamor", "Enfermedad física", "Enfermedad mental"])) &
    (df["Sexo de la victima"] == 'Hombre') &
    (df["Zona del Hecho"] == "Cabecera municipal") &
    (df["Año del hecho"] >= 2022) 
    )

df.loc[condicion_riesgo_alto, "Riesgo Alto"] = 1
df["Riesgo Alto"].value_counts()
```

### 20 casos cumplen los criterios de "Riesgo Alto"; 2481 son "Riesgo Bajo". Esto significa que la variable está desbalanceada, lo cual es muy común en problemas sociales, médicos y de seguridad.



```python
columnas_X = [
    "Ciclo Vital",
    "Sexo de la victima",
    "Escolaridad",
    "Zona del Hecho",
    "Actividad Durante el Hecho",
    "Razón del Suicidio"
]
X = df[columnas_X]
y = df["Riesgo Alto"]
X_encoded = pd.get_dummies(X, drop_first=True)
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y, test_size=0.2, random_state=42, stratify=y
)
```

## En esta sección seleccionamos las variables que se consideran relevantes para predecir el riesgo de suicidio, con base en el análisis exploratorio y el criterio clínico-psicosocial.
## Las variables seleccionadas incluyen aspectos del ciclo vital, género, escolaridad, zona del hecho, actividad realizada durante el hecho y la razón del suicidio (cuando está disponible).
## Luego se aplica codificación One-Hot (get_dummies) para convertir las variables categóricas en variables numéricas que el modelo pueda procesar.
## Finalmente, los datos se dividen en conjuntos de entrenamiento y prueba, manteniendo la proporción de clases (`stratify=y`) para garantizar que tanto riesgo alto como bajo estén representados en ambos subconjuntos.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix

# Entrenar el modelo
modelo = RandomForestClassifier(random_state=42)
modelo.fit(X_train, y_train)

# Predecir
y_pred = modelo.predict(X_test)
```

```python
print("Reporte de clasificación:")
print(classification_report(y_test, y_pred))

print("Matriz de confusión:")
print(confusion_matrix(y_test, y_pred))
```

## 📊 Evaluación del modelo

El modelo tiene una precisión general del 98%, lo cual puede parecer muy bueno a simple vista. Sin embargo, es importante observar cómo se comporta con la clase minoritaria (`Riesgo Alto = 1`), ya que el objetivo del modelo es detectar posibles casos con mayor riesgo.

- **Precisión para clase 1**: 0.14 → El modelo acierta poco cuando predice riesgo alto.
- **Recall para clase 1**: 0.25 → Solo logra identificar correctamente 1 de los 4 casos reales de riesgo alto.
- **F1-score clase 1**: 0.18 → Bajo desempeño al balancear precisión y recall para esta clase.

La **matriz de confusión** muestra que:
- 491 casos fueron correctamente clasificados como riesgo bajo.
- 6 casos fueron clasificados erróneamente como riesgo alto.
- Solo 1 caso de riesgo alto fue identificado correctamente.
- 3 casos de riesgo alto fueron ignorados por el modelo.

Este resultado evidencia una limitación importante del modelo: **la clase de interés es muy minoritaria (solo 32 casos en total)**, lo que dificulta que el modelo aprenda a reconocerla.

Este tipo de desbalance es común en problemas sociales y de salud pública, donde los casos críticos son pocos pero muy importantes. Como ejercicio inicial, el modelo permite observar qué variables son más influyentes en las decisiones del algoritmo y abrir camino para ajustes futuros (como el uso de técnicas de balanceo).


```python
import pandas as pd
import matplotlib.pyplot as plt

importancia = pd.Series(modelo.feature_importances_, index=X_encoded.columns)
importancia.sort_values(ascending=True).plot(kind='barh', figsize=(10, 8), title="Importancia de variables")
plt.tight_layout()
plt.show()
```
